```python
tiktok_usernames = tiktok_usernames.splitlines()
```

```python
import requests, json
def get_info_from_social_data_info(key, username, social_platform='tiktok', subscribe=True):
    url_base = 'https://socapi.icu/v2.0/api/reports/new/'
    headers = {"X-Api-Key": key}
    param = {
        "url": username,
        "platform": social_platform.lower(),
        "subscribe": subscribe,
    }
    response = requests.post(url_base, headers=headers, params=param)
    return json.loads(response.content.decode('utf-8'))
```

```python
key = ''
```

```python
temp = []
for i in tiktok_usernames:
    data = get_info_from_social_data_info(key, i)
    temp.append(data)
```

```python
import pandas as pd
pd.json_normalize(temp)
```
